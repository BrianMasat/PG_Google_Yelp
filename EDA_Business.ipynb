CONTINUING WITH THE EDA, NOW THAT WE HAVE OUR DATAFRAME CLEAN, WE PROCEED TO CREATE A:
 
1.) RATING DISTRIBUTION ANALYSIS

- I decided to created the Rating Distribution Analysis because we can see how the ratings (stars) are distributed in your data set. This will allow you to understand how the businesses in your dataset rank overall.

```python
#RATING DISTRIBUTION ANALYSIS:

import matplotlib.pyplot as plt 
```

```python
#Create qualifications histogram:
plt.hist(df_cleaned["stars"], bins = 5, edgecolor = "black", density=True) 

#Configurate label and title:
plt.xlabel("Qualifications (stars)")
plt.ylabel("Businesses Numbers")
plt.title("Qualification Distributions")

#Print the Histogram:
plt.show()
```

2.) CATEGORIES ANALYSIS:

- I decided to implement this one because I can explore the business categories present in your data. You can count how many businesses belong to each category and visualize the most common categories using a bar chart.

```python
#Get a list of all categories present in the DataFrame:
all_categories = df_cleaned["categories"].str.split(", ", expand = True).stack()

#Count the frequency of each category
category_counts = all_categories.value_counts()

#Select the 10 most common categories
top_categories = category_counts.head(10)

#Create a bar chart of the most common categories
plt.figure(figsize=(10, 6))
top_categories.plot(kind="bar", color="blue")
plt.xlabel("categories")
plt.ylabel("Businesses Numbers")
plt.title("Top 10 more comun categories Businesses")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

3.) ANALYSIS OF CITIES OR STATES:

- I decide this one because if you have complete information on cities and states, you could explore whether certain states or cities have a particular concentration of businesses and how that relates to ratings.

```python
#Group the data by state and calculate the average number of businesses and the average rating by state
state_stats = df_cleaned.groupby('state').agg({'business_id': 'count', 'stars': 'mean'}).reset_index()
state_stats = state_stats.rename(columns={'business_id': 'Número de Negocios', 'stars': 'Calificación Promedio'})

#Sort states by number of businesses in descending order
state_stats = state_stats.sort_values(by='Número de Negocios', ascending=False)

#Create a scatter plot to visualize the relationship between number of businesses and average rating by state
plt.figure(figsize=(10, 6))
plt.scatter(state_stats['Número de Negocios'], state_stats['Calificación Promedio'], color='blue', alpha=0.7)
plt.xlabel("Businesses Numbers")
plt.ylabel('Average mark')
plt.title('Analysis Cities or States')
plt.tight_layout()
plt.show()
```

4.) OUTLIER DETECTION:

- Picked this one because perform a ratings outlier analysis to identify if there are businesses with extremely high or low ratings compared to the average.

```python
df_cleaned 

#Calculate the GPA
average_rating = df_cleaned['stars'].mean()

#Calculate the standard deviation of grades
std_deviation = df_cleaned['stars'].std()

#Identify outliers (for example, more than 2 standard deviations)
outliers = df_cleaned[(df_cleaned['stars'] > average_rating + 2 * std_deviation) |
                      (df_cleaned['stars'] < average_rating - 2 * std_deviation)]

#Visualize outliers in a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(df_cleaned['stars'])
plt.title('Grades Boxplot')
plt.ylabel('Qualification')
plt.show()
```
